# Building Decision Tree Model and Predict

```python
# Predicting using the Decision_Tree Classifier
from sklearn.tree import DecisionTreeClassifier

model_DecisionTree = DecisionTreeClassifier(random_state = 10, criterion = "gini")

# min_samples_leaf, min_samples_split, max_depth, max_features, max_leaf_nodes

# fit the model on the data and predict the values
model_DecisionTree.fit(X_train, Y_train)
Y_pred = model_DecisionTree.predict(X_test)
print(list(zip(Y_test, Y_pred)))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
 
cfm = confusion_matrix(Y_test,Y_pred)
print(cfm)
print()
print("Classification report: ")
print(classification_report(Y_test,Y_pred))
print()
acc = accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

# Building Logistic Regression Model and Predict

```python
from sklearn.linear_model import LogisticRegression
#create a model object
classifier = LogisticRegression()
# train the model object
classifier.fit(X_train, Y_train)

# print intercept and coefficients
print(classifier.intercept_)
print(classifier.coef_)
```

```python
Y_pred = classifier.predict(X_test)
print(Y_pred)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
 
cfm = confusion_matrix(Y_test,Y_pred)
print(cfm)
 
print("Classification report: ")
 
print(classification_report(Y_test,Y_pred))
 
acc = accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

# Changing threshold for tuning and improving class 1 recall

```python
Y_pred_prob = classifier.predict_proba(X_test)
print(Y_pred_prob)
```

```python
for a in np.arange(0.4,0.61,0.01):
    predict_mine = np.where(Y_pred_prob[:,1] > a, 1, 0)
    cfm=confusion_matrix(Y_test, predict_mine)
    total_err=cfm[0,1]+cfm[1,0]
    print("Errors at threshold ", a, ":",total_err, " , type 2 error :", 
          cfm[1,0]," , type 1 error:", cfm[0,1])
```

```python
Y_pred_class_new = []
for value in Y_pred_prob[:,1]:
    if value > 0.44:
        Y_pred_class_new.append(1)
    else:
        Y_pred_class_new.append(0)
```

```python
print(Y_pred_class_new)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
 
cfm = confusion_matrix(Y_test,Y_pred_class_new)
print(cfm)
 
print("Classification report: ")
 
print(classification_report(Y_test,Y_pred_class_new))
 
acc = accuracy_score(Y_test, Y_pred_class_new)
print("Accuracy of the model: ",acc)
```
